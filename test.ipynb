```python
import numpy as np
import matplotlib.pyplot as plt

def neural_network_demo(W1, W2, W3, b1, b2, b3, activation_function):
    # Initialize network parameters
    weights = [W1, W2, W3]
    biases = [b1, b2, b3]
 
    # Function to add a new hidden layer
    def add_layer(input_size, output_size):
        W_new = np.zeros((output_size, input_size))
        b_new = np.zeros((output_size, 1))
        weights.insert(-1, W_new)
        biases.insert(-1, b_new)
    
    # Function to add neurons to an existing layer
    def add_neurons(layer_index, num_neurons):
        input_size = weights[layer_index].shape[1]
        W_extra = np.zeros((num_neurons, input_size))
        b_extra = np.zeros((num_neurons, 1))
        weights[layer_index] = np.vstack([weights[layer_index], W_extra])
        biases[layer_index] = np.vstack([biases[layer_index], b_extra])
        # Adjust weights of the next layer
        weights[layer_index + 1] = np.hstack([weights[layer_index + 1], np.zeros((weights[layer_index + 1].shape[0], num_neurons))])
    
    # Forward pass function
    def forward(x):
        a = x.reshape(-1, 1)  # Convert to column vector
        for i in range(len(weights)):
            z = weights[i] @ a + biases[i]
            if i < len(weights) - 1:
                a = activation_function(z)
            else:
                a =  z
        print(a)
        return a
    
    
    # Perform grid search
    grid_size = 100  # Adjust grid size as needed
    x1 = np.linspace(-10, 10, grid_size)
    x2 = np.linspace(-10, 10, grid_size)
    X1, X2 = np.meshgrid(x1, x2)
    Y = np.zeros_like(X1)

    
    for i in range(grid_size):
        for j in range(grid_size):
            input_vec = np.array([X1[i, j], X2[i, j]])
            hidden_layer_output = forward(input_vec)
            Y[i, j] = hidden_layer_output
       
    
    # Plot the color contour
    plt.figure(figsize=(10, 8))
    contour = plt.contourf(X1, X2, Y , levels=100, cmap='viridis', alpha=0.8)
    plt.colorbar(contour, label='Output Scalar')
    
    # Overlay scatter plot with numerical annotations every 5 steps
    step = 5  # Adjust step to control density of annotations
    for i in range(0, grid_size, step):
        for j in range(0, grid_size, step):
            #plt.scatter(X1[i, j], X2[i, j], color='white', edgecolors='black', s=50)
            plt.text(X1[i, j], X2[i, j], f"{Y[i, j]:.2f}", color='black',
                     fontsize=8, ha='center', va='center')
    
    plt.xlabel('Input 1')
    plt.ylabel('Input 2')
    plt.title('Neural Network Output with Annotations')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
    
    # Return parameters and utility functions for manual modification
    return {
        'weights': weights,
        'biases': biases,
        'add_layer': add_layer,
        'add_neurons': add_neurons,
        'forward': forward,
        'activation_function': activation_function
    }

# Define the activation function (ReLU)
def activation_function(x):
    return np.abs(x)
    #return np.maximum(0, x)
    # To use a different activation function, modify the above line or define a new function.

# Run the neural network demo multiple times
for run in range(1):
    # Initialize weights and biases with random values
    W1 = np.random.randn(2, 2)   
    b1 = np.random.randn(2, 1)
    W2 = np.random.randn(2, 2)
    b2 = np.random.randn(2, 1)
    W3 = np.random.randn(1, 2)
    b3 = np.random.randn(1, 1)
    
    print(f"Run {run + 1}/10")
    params = neural_network_demo(W1=W1, W2=W2, W3=W3, b2=b2, b3=b3, b1=b1, activation_function=activation_function)
    
    # Example of manually modifying weights and biases
    # Uncomment and modify as needed
    # params['weights'][0][0,0] = 1.0  # Modify first weight of first layer
    # params['biases'][2][0,0] = -0.5  # Modify bias of output layer
    
    # You can also add layers or neurons if needed
    # params['add_layer'](input_size=4, output_size=2)  # Add a new hidden layer with 2 neurons
    # params['add_neurons'](layer_index=0, num_neurons=2)  # Add 2 neurons to the first hidden layer
    
    print("Plot generated.\n")
```
